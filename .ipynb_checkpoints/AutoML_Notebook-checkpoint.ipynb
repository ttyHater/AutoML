# AutoML Notebook: Generalized Machine Learning Pipeline

This notebook provides a unified interface for:
- **Supervised Learning Models**: kNN Classification/Regression, Linear Models (Regression/Classification)
- **Curve Fitting**: Linear, Quadratic, Cubic, Sine, Exponential, Power Law, Gaussian, Logistic
- **Cross-Validation Strategies**: K-Fold, 3-Way Fold, Monte Carlo
- **Optional Visualizations**: All graphs can be toggled on/off

## Features:
- Flexible model selection
- Multiple cross-validation methods
- Automatic model evaluation and comparison
- Unified parameter tuning
- Optional visualization for all outputs

````python
# Import libraries
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.model_selection import cross_val_score, cross_validate, KFold, ShuffleSplit
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    mean_squared_error, r2_score, mean_absolute_error
)
from sklearn.pipeline import Pipeline
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)
````

## Section 1: Cross-Validation Strategies

````python
class CrossValidationManager:
    """
    Unified interface for multiple cross-validation strategies.
    """
    
    @staticmethod
    def kfold(X, y, model, n_splits=5, scoring=None):
        """
        K-Fold Cross-Validation.
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Feature matrix
        y : array-like, shape (n_samples,)
            Target variable
        model : estimator object
            Scikit-learn compatible model
        n_splits : int, default=5
            Number of folds
        scoring : str or callable, optional
            Scoring metric
        
        Returns:
        --------
        dict : Results containing scores and fold indices
        """
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
        scores = cross_val_score(model, X, y, cv=kf, scoring=scoring)
        
        return {
            'method': 'K-Fold',
            'n_splits': n_splits,
            'scores': scores,
            'mean': scores.mean(),
            'std': scores.std(),
            'cv_object': kf
        }
    
    @staticmethod
    def three_fold(X, y, model, scoring=None):
        """
        3-Way Fold Cross-Validation (specialized K-Fold with k=3).
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Feature matrix
        y : array-like, shape (n_samples,)
            Target variable
        model : estimator object
            Scikit-learn compatible model
        scoring : str or callable, optional
            Scoring metric
        
        Returns:
        --------
        dict : Results containing scores and fold information
        """
        return CrossValidationManager.kfold(X, y, model, n_splits=3, scoring=scoring)
    
    @staticmethod
    def monte_carlo(X, y, model, n_iterations=10, test_size=0.2, scoring=None):
        """
        Monte Carlo Cross-Validation (repeated random train/test splits).
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Feature matrix
        y : array-like, shape (n_samples,)
            Target variable
        model : estimator object
            Scikit-learn compatible model
        n_iterations : int, default=10
            Number of random train/test splits
        test_size : float, default=0.2
            Proportion of dataset to include in test split
        scoring : str or callable, optional
            Scoring metric
        
        Returns:
        --------
        dict : Results containing scores from multiple iterations
        """
        mccv = ShuffleSplit(n_splits=n_iterations, test_size=test_size, random_state=42)
        scores = cross_val_score(model, X, y, cv=mccv, scoring=scoring)
        
        return {
            'method': 'Monte Carlo',
            'n_iterations': n_iterations,
            'test_size': test_size,
            'scores': scores,
            'mean': scores.mean(),
            'std': scores.std(),
            'cv_object': mccv
        }
````

## Section 2: Supervised Learning Models

````python
class SupervisedLearningPipeline:
    """
    Unified interface for supervised learning models with integrated cross-validation.
    """
    
    def __init__(self, model_type, **kwargs):
        """
        Initialize model.
        
        Parameters:
        -----------
        model_type : str
            Type of model: 'knn_classifier', 'knn_regressor', 'linear_regression', 'logistic_regression'
        **kwargs : dict
            Model-specific hyperparameters
        """
        self.model_type = model_type
        self.model = self._initialize_model(**kwargs)
        self.is_classifier = 'classifier' in model_type.lower()
    
    def _initialize_model(self, **kwargs):
        """Initialize model based on type."""
        if self.model_type == 'knn_classifier':
            return KNeighborsClassifier(**kwargs)
        elif self.model_type == 'knn_regressor':
            return KNeighborsRegressor(**kwargs)
        elif self.model_type == 'linear_regression':
            return LinearRegression(**kwargs)
        elif self.model_type == 'logistic_regression':
            return LogisticRegression(max_iter=1000, random_state=42, **kwargs)
        else:
            raise ValueError(f"Unknown model type: {self.model_type}")
    
    def fit(self, X, y):
        """Fit model to training data."""
        self.model.fit(X, y)
        return self
    
    def predict(self, X):
        """Make predictions."""
        return self.model.predict(X)
    
    def evaluate(self, X, y):
        """Evaluate model performance."""
        y_pred = self.predict(X)
        
        results = {}
        
        if self.is_classifier:
            results['accuracy'] = accuracy_score(y, y_pred)
            results['precision'] = precision_score(y, y_pred, average='weighted', zero_division=0)
            results['recall'] = recall_score(y, y_pred, average='weighted', zero_division=0)
            results['f1'] = f1_score(y, y_pred, average='weighted', zero_division=0)
        else:
            results['mse'] = mean_squared_error(y, y_pred)
            results['rmse'] = np.sqrt(results['mse'])
            results['mae'] = mean_absolute_error(y, y_pred)
            results['r2'] = r2_score(y, y_pred)
        
        return results
    
    def cross_validate(self, X, y, method='kfold', show_plot=False, **cv_kwargs):
        """
        Perform cross-validation using specified method.
        
        Parameters:
        -----------
        X : array-like
            Feature matrix
        y : array-like
            Target variable
        method : str, default='kfold'
            Cross-validation method: 'kfold', 'three_fold', 'monte_carlo'
        show_plot : bool, default=False
            Whether to display visualization of results
        **cv_kwargs : dict
            Additional arguments for cross-validation method
        
        Returns:
        --------
        dict : Cross-validation results
        """
        cv_manager = CrossValidationManager()
        
        # Choose scoring metric based on model type
        if self.is_classifier:
            scoring = cv_kwargs.pop('scoring', 'accuracy')
        else:
            scoring = cv_kwargs.pop('scoring', 'r2')
        
        # Perform cross-validation
        if method == 'kfold':
            n_splits = cv_kwargs.pop('n_splits', 5)
            results = cv_manager.kfold(X, y, self.model, n_splits=n_splits, scoring=scoring)
        elif method == 'three_fold':
            results = cv_manager.three_fold(X, y, self.model, scoring=scoring)
        elif method == 'monte_carlo':
            n_iterations = cv_kwargs.pop('n_iterations', 10)
            test_size = cv_kwargs.pop('test_size', 0.2)
            results = cv_manager.monte_carlo(X, y, self.model, n_iterations=n_iterations, 
                                             test_size=test_size, scoring=scoring)
        else:
            raise ValueError(f"Unknown CV method: {method}")
        
        # Print results
        print(f"\n{'='*60}")
        print(f"Cross-Validation Results: {results['method']}")
        print(f"{'='*60}")
        print(f"Model: {self.model_type}")
        print(f"Mean Score: {results['mean']:.4f} (+/- {results['std']:.4f})")
        print(f"Individual Scores: {np.round(results['scores'], 4)}")
        
        # Optional visualization
        if show_plot:
            self._plot_cv_results(results)
        
        return results
    
    def _plot_cv_results(self, results):
        """Plot cross-validation results."""
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        
        # Bar plot of fold scores
        axes[0].bar(range(len(results['scores'])), results['scores'], alpha=0.7, color='steelblue')
        axes[0].axhline(y=results['mean'], color='r', linestyle='--', label=f"Mean: {results['mean']:.4f}")
        axes[0].set_xlabel('Fold')
        axes[0].set_ylabel('Score')
        axes[0].set_title(f"{results['method']} CV Scores")
        axes[0].legend()
        axes[0].grid(axis='y', alpha=0.3)
        
        # Box plot
        axes[1].boxplot(results['scores'], vert=True)
        axes[1].set_ylabel('Score')
        axes[1].set_title('Score Distribution')
        axes[1].grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        plt.show()
````

## Section 3: Curve Fitting

````python
class CurveFittingToolkit:
    """
    Collection of parametric curve fitting functions with unified interface.
    """
    
    # Define curve functions
    @staticmethod
    def linear(x, m, b):
        return m * x + b
    
    @staticmethod
    def quadratic(x, a, b, c):
        return a * x**2 + b * x + c
    
    @staticmethod
    def cubic(x, a, b, c, d):
        return a * x**3 + b * x**2 + c * x + d
    
    @staticmethod
    def sine(x, A, f, C, D):
        return A * np.sin((2 * np.pi * f * x) + C) + D
    
    @staticmethod
    def exponential(x, A, b):
        return A * np.exp(-b * x)
    
    @staticmethod
    def powerlaw(x, A, alpha):
        return A * (np.power(x, -alpha))
    
    @staticmethod
    def gaussian(x, A, mu, sigma):
        return A * np.exp(-((x - mu)**2) / (2 * sigma**2))
    
    @staticmethod
    def logistic(x, L, k, x0):
        return L / (1 + np.exp(-k * (x - x0)))
    
    @staticmethod
    def sigmoid(x, A, x0, k):
        return A / (1 + np.exp(-k * (x - x0)))
    
    # Dictionary for easy access
    FUNCTIONS = {
        'linear': (linear, ['m', 'b']),
        'quadratic': (quadratic, ['a', 'b', 'c']),
        'cubic': (cubic, ['a', 'b', 'c', 'd']),
        'sine': (sine, ['A', 'f', 'C', 'D']),
        'exponential': (exponential, ['A', 'b']),
        'powerlaw': (powerlaw, ['A', 'alpha']),
        'gaussian': (gaussian, ['A', 'mu', 'sigma']),
        'logistic': (logistic, ['L', 'k', 'x0']),
        'sigmoid': (sigmoid, ['A', 'x0', 'k'])
    }
````

````python
class CurveFitter:
    """
    Unified interface for fitting parametric curves to data.
    """
    
    def __init__(self, curve_type, initial_guess=None):
        """
        Initialize curve fitter.
        
        Parameters:
        -----------
        curve_type : str
            Type of curve: 'linear', 'quadratic', 'cubic', 'sine', 'exponential', 
            'powerlaw', 'gaussian', 'logistic', 'sigmoid'
        initial_guess : array-like, optional
            Initial parameter estimates
        """
        if curve_type not in CurveFittingToolkit.FUNCTIONS:
            raise ValueError(f"Unknown curve type: {curve_type}. Available: {list(CurveFittingToolkit.FUNCTIONS.keys())}")
        
        self.curve_type = curve_type
        self.func, self.param_names = CurveFittingToolkit.FUNCTIONS[curve_type]
        self.initial_guess = initial_guess
        self.popt = None
        self.pcov = None
    
    def fit(self, x, y, **curve_fit_kwargs):
        """
        Fit curve to data.
        
        Parameters:
        -----------
        x : array-like
            Independent variable
        y : array-like
            Dependent variable
        **curve_fit_kwargs : dict
            Additional arguments for scipy.optimize.curve_fit
        
        Returns:
        --------
        self : CurveFitter object
        """
        try:
            self.popt, self.pcov = curve_fit(self.func, x, y, p0=self.initial_guess, **curve_fit_kwargs)
            return self
        except Exception as e:
            print(f"Error fitting curve: {e}")
            return None
    
    def predict(self, x):
        """Make predictions using fitted curve."""
        if self.popt is None:
            raise ValueError("Model not fitted. Call fit() first.")
        return self.func(x, *self.popt)
    
    def calculate_ssr(self, x, y):
        """Calculate Sum of Squared Residuals."""
        y_pred = self.predict(x)
        return np.sum((y - y_pred)**2)
    
    def calculate_r2(self, x, y):
        """Calculate R-squared."""
        ss_res = self.calculate_ssr(x, y)
        ss_tot = np.sum((y - np.mean(y))**2)
        return 1 - (ss_res / ss_tot)
    
    def evaluate(self, x, y):
        """Evaluate fit quality."""
        ssr = self.calculate_ssr(x, y)
        r2 = self.calculate_r2(x, y)
        rmse = np.sqrt(ssr / len(x))
        
        return {
            'ssr': ssr,
            'rmse': rmse,
            'r2': r2
        }
    
    def print_results(self, x, y):
        """Print fitting results."""
        if self.popt is None:
            print("Model not fitted.")
            return
        
        print(f"\n{'='*60}")
        print(f"Curve Fit Results: {self.curve_type.upper()}")
        print(f"{'='*60}")
        
        # Print parameters
        for name, value in zip(self.param_names, self.popt):
            print(f"{name:>10} = {value:>12.6f}")
        
        # Print metrics
        metrics = self.evaluate(x, y)
        print(f"\nMetrics:")
        print(f"SSR  = {metrics['ssr']:>12.6f}")
        print(f"RMSE = {metrics['rmse']:>12.6f}")
        print(f"R²   = {metrics['r2']:>12.6f}")
    
    def plot(self, x, y, x_range=None, title=None, xlabel='x', ylabel='y'):
        """Plot data and fitted curve."""
        if self.popt is None:
            print("Model not fitted. Call fit() first.")
            return
        
        if x_range is None:
            x_range = np.linspace(np.min(x), np.max(x), 300)
        
        y_fit = self.predict(x_range)
        
        plt.figure(figsize=(10, 6))
        plt.scatter(x, y, alpha=0.6, s=50, label='Data', color='steelblue')
        plt.plot(x_range, y_fit, 'r-', lw=2.5, label=f'Fitted {self.curve_type.capitalize()}')
        
        plt.xlabel(xlabel, fontsize=12)
        plt.ylabel(ylabel, fontsize=12)
        plt.title(title or f'{self.curve_type.capitalize()} Curve Fit', fontsize=14)
        plt.legend(fontsize=11)
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
````

## Section 4: Model Comparison and Selection

````python
class ModelComparator:
    """
    Compare multiple models using the same cross-validation strategy.
    """
    
    def __init__(self):
        self.results = {}
        self.models = {}
    
    def add_model(self, name, model):
        """
        Add model to comparison.
        
        Parameters:
        -----------
        name : str
            Model name/identifier
        model : SupervisedLearningPipeline
            Model to add
        """
        self.models[name] = model
    
    def compare(self, X, y, cv_method='kfold', show_plot=False, **cv_kwargs):
        """
        Compare all added models using specified CV method.
        
        Parameters:
        -----------
        X : array-like
            Feature matrix
        y : array-like
            Target variable
        cv_method : str, default='kfold'
            Cross-validation method
        show_plot : bool, default=False
            Whether to display comparison plot
        **cv_kwargs : dict
            Additional CV arguments
        
        Returns:
        --------
        pd.DataFrame : Comparison results
        """
        print(f"\nComparing {len(self.models)} models using {cv_method} cross-validation...\n")
        
        comparison_data = []
        
        for name, model in self.models.items():
            results = model.cross_validate(X, y, method=cv_method, show_plot=False, **cv_kwargs)
            
            comparison_data.append({
                'Model': name,
                'Mean Score': results['mean'],
                'Std Dev': results['std'],
                'Min Score': results['scores'].min(),
                'Max Score': results['scores'].max()
            })
            
            self.results[name] = results
        
        comparison_df = pd.DataFrame(comparison_data).sort_values('Mean Score', ascending=False)
        
        print(comparison_df.to_string(index=False))
        print(f"\nBest Model: {comparison_df.iloc[0]['Model']} (Score: {comparison_df.iloc[0]['Mean Score']:.4f})")
        
        if show_plot:
            self._plot_comparison(comparison_df)
        
        return comparison_df
    
    def _plot_comparison(self, df):
        """Plot model comparison."""
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        
        # Bar plot with error bars
        axes[0].barh(df['Model'], df['Mean Score'], xerr=df['Std Dev'], 
                     alpha=0.7, capsize=5, color='steelblue')
        axes[0].set_xlabel('Mean CV Score')
        axes[0].set_title('Model Comparison: Mean Scores')
        axes[0].grid(axis='x', alpha=0.3)
        
        # Score range
        axes[1].scatter(df['Mean Score'], df['Model'], s=100, alpha=0.7, color='steelblue')
        for idx, row in df.iterrows():
            axes[1].plot([row['Min Score'], row['Max Score']], [row['Model'], row['Model']], 
                       'k-', alpha=0.5, lw=2)
        axes[1].set_xlabel('Score Range')
        axes[1].set_title('Score Range Across Folds')
        axes[1].grid(axis='x', alpha=0.3)
        
        plt.tight_layout()
        plt.show()
````

## Section 5: Example Usage

### Example 1: Supervised Learning with Cross-Validation

This example demonstrates how to use the SupervisedLearningPipeline with different cross-validation methods.

````python
# Generate synthetic classification dataset
from sklearn.datasets import make_classification, make_regression

print("Generating synthetic classification dataset...")
X_class, y_class = make_classification(n_samples=150, n_features=10, n_informative=5, 
                                        n_redundant=2, random_state=42)

# Initialize and evaluate KNN Classifier with different CV methods
knn_clf = SupervisedLearningPipeline('knn_classifier', n_neighbors=5)

print("\n" + "="*60)
print("KNN Classifier - 3-Way Fold Cross-Validation")
print("="*60)
results_3fold = knn_clf.cross_validate(X_class, y_class, method='three_fold', show_plot=True)

print("\n" + "="*60)
print("KNN Classifier - K-Fold Cross-Validation (k=5)")
print("="*60)
results_kfold = knn_clf.cross_validate(X_class, y_class, method='kfold', 
                                       n_splits=5, show_plot=True)

print("\n" + "="*60)
print("KNN Classifier - Monte Carlo Cross-Validation")
print("="*60)
results_mc = knn_clf.cross_validate(X_class, y_class, method='monte_carlo', 
                                    n_iterations=10, test_size=0.2, show_plot=True)
````

### Example 2: Regression with Multiple CV Methods

````python
# Generate synthetic regression dataset
print("\nGenerating synthetic regression dataset...")
X_reg, y_reg = make_regression(n_samples=150, n_features=10, n_informative=5, random_state=42)

# Initialize and evaluate KNN Regressor
knn_reg = SupervisedLearningPipeline('knn_regressor', n_neighbors=5)

print("\n" + "="*60)
print("KNN Regressor - 3-Way Fold Cross-Validation")
print("="*60)
results_reg_3fold = knn_reg.cross_validate(X_reg, y_reg, method='three_fold', show_plot=False)

print("\n" + "="*60)
print("KNN Regressor - Monte Carlo Cross-Validation")
print("="*60)
results_reg_mc = knn_reg.cross_validate(X_reg, y_reg, method='monte_carlo', 
                                        n_iterations=10, show_plot=False)
````

### Example 3: Model Comparison

````python
# Compare multiple classification models
print("\nComparing multiple classification models...\n")

comparator = ModelComparator()
comparator.add_model('KNN (k=3)', SupervisedLearningPipeline('knn_classifier', n_neighbors=3))
comparator.add_model('KNN (k=5)', SupervisedLearningPipeline('knn_classifier', n_neighbors=5))
comparator.add_model('KNN (k=7)', SupervisedLearningPipeline('knn_classifier', n_neighbors=7))
comparator.add_model('Logistic Regression', SupervisedLearningPipeline('logistic_regression'))

comparison_results = comparator.compare(X_class, y_class, cv_method='kfold', 
                                        n_splits=5, show_plot=True)
````

### Example 4: Curve Fitting

````python
# Generate synthetic data with underlying curve
np.random.seed(42)
x_curve = np.linspace(0, 10, 50)
y_curve = 2.5 * x_curve + 3 + np.random.normal(0, 2, len(x_curve))  # Linear with noise

# Fit linear curve
fitter_linear = CurveFitter('linear')
fitter_linear.fit(x_curve, y_curve)
fitter_linear.print_results(x_curve, y_curve)
fitter_linear.plot(x_curve, y_curve, title='Linear Curve Fit')

# Generate quadratic data
y_quad = 0.5 * x_curve**2 - 2 * x_curve + 10 + np.random.normal(0, 3, len(x_curve))

fitter_quad = CurveFitter('quadratic')
fitter_quad.fit(x_curve, y_quad)
fitter_quad.print_results(x_curve, y_quad)
fitter_quad.plot(x_curve, y_quad, title='Quadratic Curve Fit')

# Generate sinusoidal data
y_sine = 5 * np.sin(0.5 * x_curve) + 10 + np.random.normal(0, 0.5, len(x_curve))

fitter_sine = CurveFitter('sine', initial_guess=[5, 0.5, 0, 10])
fitter_sine.fit(x_curve, y_sine)
fitter_sine.print_results(x_curve, y_sine)
fitter_sine.plot(x_curve, y_sine, title='Sine Curve Fit')
````

### Example 5: Compare Multiple Curve Fits

````python
# Fit multiple curves to the same data
x_test = np.linspace(0, 10, 50)
y_test = 0.3 * x_test**2 + np.random.normal(0, 2, len(x_test))

curve_types = ['linear', 'quadratic', 'cubic']
results_curves = []

for curve_type in curve_types:
    fitter = CurveFitter(curve_type)
    fitter.fit(x_test, y_test)
    metrics = fitter.evaluate(x_test, y_test)
    
    results_curves.append({
        'Curve Type': curve_type.capitalize(),
        'SSR': metrics['ssr'],
        'RMSE': metrics['rmse'],
        'R²': metrics['r2']
    })

curve_comparison = pd.DataFrame(results_curves).sort_values('R²', ascending=False)
print("\nCurve Fitting Comparison:")
print(curve_comparison.to_string(index=False))
````

## Section 6: Quick Reference Guide

### SupervisedLearningPipeline Quick Start

```python
# Initialize model
model = SupervisedLearningPipeline('knn_classifier', n_neighbors=5)

# Fit model
model.fit(X_train, y_train)

# Evaluate on test data
metrics = model.evaluate(X_test, y_test)

# Cross-validate with visualization
cv_results = model.cross_validate(X, y, method='kfold', n_splits=5, show_plot=True)
cv_results = model.cross_validate(X, y, method='three_fold', show_plot=False)
cv_results = model.cross_validate(X, y, method='monte_carlo', n_iterations=10, show_plot=True)
```

### Supported Models
- `'knn_classifier'`: k-Nearest Neighbors Classification
- `'knn_regressor'`: k-Nearest Neighbors Regression
- `'linear_regression'`: Linear Regression
- `'logistic_regression'`: Logistic Regression

### CurveFitter Quick Start

```python
# Initialize and fit
fitter = CurveFitter('quadratic')
fitter.fit(x, y)

# Make predictions
y_pred = fitter.predict(x_new)

# Evaluate
metrics = fitter.evaluate(x, y)

# Visualize (optional)
fitter.plot(x, y)
fitter.print_results(x, y)
```

### Supported Curves
- `'linear'`: y = mx + b
- `'quadratic'`: y = ax² + bx + c
- `'cubic'`: y = ax³ + bx² + cx + d
- `'sine'`: y = A·sin(2πfx + C) + D
- `'exponential'`: y = A·exp(-bx)
- `'powerlaw'`: y = A·x^(-α)
- `'gaussian'`: y = A·exp(-(x-μ)²/(2σ²))
- `'logistic'`: y = L/(1 + exp(-k(x-x₀)))
- `'sigmoid'`: y = A/(1 + exp(-k(x-x₀)))

### Cross-Validation Methods
- `'kfold'`: Standard K-Fold (default k=5)
- `'three_fold'`: 3-Way Fold (K-Fold with k=3)
- `'monte_carlo'`: Repeated random train/test splits

### Visualization Control
All methods accept `show_plot=True/False` parameter to control visualization.
