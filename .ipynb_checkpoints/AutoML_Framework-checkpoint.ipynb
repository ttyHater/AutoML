# AutoML Framework
## Generalized Machine Learning with Multiple Cross-Validation Strategies

This notebook provides two main functions:
- `automl_classifier()` - For classification tasks
- `automl_regressor()` - For regression tasks

Both support multiple cross-validation methods:
- **3-way fold**: Simple 3-fold cross-validation
- **k-fold**: Standard k-fold cross-validation (default k=5)
- **Monte Carlo**: Multiple random train-test splits

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,
    r2_score, mean_squared_error, mean_absolute_error
)
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## Classification Function

```python
def automl_classifier(X, y, model, 
                     cv_method='kfold', 
                     cv_params=None,
                     test_size=0.25,
                     scale_features=True,
                     random_state=42,
                     plot=False):
    """
    Automated Machine Learning for Classification with Multiple Cross-Validation Methods
    
    Parameters:
    -----------
    X : array-like or DataFrame
        Feature matrix
    y : array-like or Series
        Target variable
    model : sklearn classifier object
        Instantiated classifier model (e.g., KNeighborsClassifier(), LogisticRegression())
    cv_method : str, default='kfold'
        Cross-validation method: 'kfold', '3way', or 'montecarlo'
    cv_params : dict, default=None
        Parameters for cross-validation:
        - 'kfold': {'n_splits': int} (default: 5)
        - 'montecarlo': {'n_iterations': int, 'test_size': float} (default: 100, 0.25)
    test_size : float, default=0.25
        Proportion of data for testing (used if cv_method='montecarlo')
    scale_features : bool, default=True
        Whether to standardize features
    random_state : int, default=42
        Random state for reproducibility
    plot : bool, default=False
        Whether to generate visualizations
    
    Returns:
    --------
    results : dict
        Dictionary containing:
        - 'cv_method': Cross-validation method used
        - 'scores': Array of cross-validation scores
        - 'mean_score': Mean cross-validation score
        - 'std_score': Standard deviation of scores
        - 'trained_model': Fitted model on full dataset
        - 'metrics': Dict of additional metrics (for montecarlo only)
    """
    
    # Set default cv_params
    if cv_params is None:
        cv_params = {}
    
    # Scale features if requested
    if scale_features:
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
    
    results = {'cv_method': cv_method}
    
    # === 3-Way Fold Cross-Validation ===
    if cv_method == '3way':
        kf = KFold(n_splits=3, shuffle=True, random_state=random_state)
        scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
        
        results['scores'] = scores
        results['mean_score'] = scores.mean()
        results['std_score'] = scores.std()
        results['folds'] = 3
        
        print(f"3-Way Fold Cross-Validation Results:")
        print(f"  Fold Scores: {scores}")
        print(f"  Mean Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    # === K-Fold Cross-Validation ===
    elif cv_method == 'kfold':
        n_splits = cv_params.get('n_splits', 5)
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
        scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
        
        results['scores'] = scores
        results['mean_score'] = scores.mean()
        results['std_score'] = scores.std()
        results['folds'] = n_splits
        
        print(f"{n_splits}-Fold Cross-Validation Results:")
        print(f"  Fold Scores: {scores}")
        print(f"  Mean Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    # === Monte Carlo Cross-Validation ===
    elif cv_method == 'montecarlo':
        n_iterations = cv_params.get('n_iterations', 100)
        test_sz = cv_params.get('test_size', test_size)
        
        train_scores = []
        test_scores = []
        metrics_list = []
        
        for i in range(n_iterations):
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=test_sz, random_state=random_state + i
            )
            
            model.fit(X_train, y_train)
            train_scores.append(model.score(X_train, y_train))
            test_scores.append(model.score(X_test, y_test))
            
            # Calculate additional metrics on the last iteration
            if i == n_iterations - 1:
                y_pred = model.predict(X_test)
                metrics_list = {
                    'accuracy': accuracy_score(y_test, y_pred),
                    'precision': precision_score(y_test, y_pred, average='weighted', zero_division=0),
                    'recall': recall_score(y_test, y_pred, average='weighted', zero_division=0),
                    'f1': f1_score(y_test, y_pred, average='weighted', zero_division=0)
                }
        
        results['train_scores'] = np.array(train_scores)
        results['test_scores'] = np.array(test_scores)
        results['mean_train_score'] = np.mean(train_scores)
        results['mean_test_score'] = np.mean(test_scores)
        results['std_train_score'] = np.std(train_scores)
        results['std_test_score'] = np.std(test_scores)
        results['iterations'] = n_iterations
        results['metrics'] = metrics_list
        
        print(f"Monte Carlo Cross-Validation Results ({n_iterations} iterations):")
        print(f"  Mean Train Accuracy: {np.mean(train_scores):.4f} (+/- {np.std(train_scores):.4f})")
        print(f"  Mean Test Accuracy: {np.mean(test_scores):.4f} (+/- {np.std(test_scores):.4f})")
        print(f"  Precision: {metrics_list['precision']:.4f}")
        print(f"  Recall: {metrics_list['recall']:.4f}")
        print(f"  F1-Score: {metrics_list['f1']:.4f}")
    
    else:
        raise ValueError(f"cv_method must be '3way', 'kfold', or 'montecarlo', got {cv_method}")
    
    # Train final model on entire dataset
    model.fit(X, y)
    results['trained_model'] = model
    
    # Generate plots if requested
    if plot:
        _plot_classifier_results(results, cv_method)
    
    return results
```

## Regression Function

```python
def automl_regressor(X, y, model,
                    cv_method='kfold',
                    cv_params=None,
                    test_size=0.25,
                    scale_features=True,
                    random_state=42,
                    plot=False):
    """
    Automated Machine Learning for Regression with Multiple Cross-Validation Methods
    
    Parameters:
    -----------
    X : array-like or DataFrame
        Feature matrix
    y : array-like or Series
        Target variable
    model : sklearn regressor object
        Instantiated regressor model (e.g., KNeighborsRegressor(), LinearRegression())
    cv_method : str, default='kfold'
        Cross-validation method: 'kfold', '3way', or 'montecarlo'
    cv_params : dict, default=None
        Parameters for cross-validation:
        - 'kfold': {'n_splits': int} (default: 5)
        - 'montecarlo': {'n_iterations': int, 'test_size': float} (default: 100, 0.25)
    test_size : float, default=0.25
        Proportion of data for testing (used if cv_method='montecarlo')
    scale_features : bool, default=True
        Whether to standardize features
    random_state : int, default=42
        Random state for reproducibility
    plot : bool, default=False
        Whether to generate visualizations
    
    Returns:
    --------
    results : dict
        Dictionary containing:
        - 'cv_method': Cross-validation method used
        - 'scores': Array of cross-validation R² scores
        - 'mean_score': Mean R² score
        - 'std_score': Standard deviation of R² scores
        - 'trained_model': Fitted model on full dataset
        - 'metrics': Dict of additional metrics (for montecarlo only)
    """
    
    # Set default cv_params
    if cv_params is None:
        cv_params = {}
    
    # Scale features if requested
    if scale_features:
        scaler = StandardScaler()
        X = scaler.fit_transform(X)
    
    results = {'cv_method': cv_method}
    
    # === 3-Way Fold Cross-Validation ===
    if cv_method == '3way':
        kf = KFold(n_splits=3, shuffle=True, random_state=random_state)
        scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
        
        results['scores'] = scores
        results['mean_score'] = scores.mean()
        results['std_score'] = scores.std()
        results['folds'] = 3
        
        print(f"3-Way Fold Cross-Validation Results:")
        print(f"  Fold R² Scores: {scores}")
        print(f"  Mean R²: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    # === K-Fold Cross-Validation ===
    elif cv_method == 'kfold':
        n_splits = cv_params.get('n_splits', 5)
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=random_state)
        scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
        
        results['scores'] = scores
        results['mean_score'] = scores.mean()
        results['std_score'] = scores.std()
        results['folds'] = n_splits
        
        print(f"{n_splits}-Fold Cross-Validation Results:")
        print(f"  Fold R² Scores: {scores}")
        print(f"  Mean R²: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    # === Monte Carlo Cross-Validation ===
    elif cv_method == 'montecarlo':
        n_iterations = cv_params.get('n_iterations', 100)
        test_sz = cv_params.get('test_size', test_size)
        
        train_scores = []
        test_scores = []
        mse_scores = []
        mae_scores = []
        
        for i in range(n_iterations):
            X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=test_sz, random_state=random_state + i
            )
            
            model.fit(X_train, y_train)
            train_scores.append(model.score(X_train, y_train))
            test_scores.append(model.score(X_test, y_test))
            
            # Calculate additional metrics on the last iteration
            if i == n_iterations - 1:
                y_pred = model.predict(X_test)
                mse_scores = mean_squared_error(y_test, y_pred)
                mae_scores = mean_absolute_error(y_test, y_pred)
        
        results['train_scores'] = np.array(train_scores)
        results['test_scores'] = np.array(test_scores)
        results['mean_train_score'] = np.mean(train_scores)
        results['mean_test_score'] = np.mean(test_scores)
        results['std_train_score'] = np.std(train_scores)
        results['std_test_score'] = np.std(test_scores)
        results['iterations'] = n_iterations
        results['metrics'] = {
            'mse': mse_scores,
            'mae': mae_scores,
            'rmse': np.sqrt(mse_scores)
        }
        
        print(f"Monte Carlo Cross-Validation Results ({n_iterations} iterations):")
        print(f"  Mean Train R²: {np.mean(train_scores):.4f} (+/- {np.std(train_scores):.4f})")
        print(f"  Mean Test R²: {np.mean(test_scores):.4f} (+/- {np.std(test_scores):.4f})")
        print(f"  MSE: {mse_scores:.4f}")
        print(f"  MAE: {mae_scores:.4f}")
        print(f"  RMSE: {np.sqrt(mse_scores):.4f}")
    
    else:
        raise ValueError(f"cv_method must be '3way', 'kfold', or 'montecarlo', got {cv_method}")
    
    # Train final model on entire dataset
    model.fit(X, y)
    results['trained_model'] = model
    
    # Generate plots if requested
    if plot:
        _plot_regressor_results(results, cv_method)
    
    return results
```

## Helper Functions for Visualization

```python
def _plot_classifier_results(results, cv_method):
    """
    Generate visualizations for classifier results
    """
    if cv_method in ['3way', 'kfold']:
        fig, ax = plt.subplots(figsize=(10, 6))
        folds = list(range(1, len(results['scores']) + 1))
        ax.bar(folds, results['scores'], alpha=0.7, color='steelblue', edgecolor='black')
        ax.axhline(y=results['mean_score'], color='red', linestyle='--', linewidth=2, label=f"Mean: {results['mean_score']:.4f}")
        ax.fill_between(folds, 
                         results['mean_score'] - results['std_score'],
                         results['mean_score'] + results['std_score'],
                         alpha=0.2, color='red')
        ax.set_xlabel('Fold', fontsize=12)
        ax.set_ylabel('Accuracy', fontsize=12)
        ax.set_title(f'Classification: {cv_method.upper()} Cross-Validation', fontsize=14, fontweight='bold')
        ax.set_xticks(folds)
        ax.legend(fontsize=10)
        ax.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    elif cv_method == 'montecarlo':
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        
        # Plot 1: Train vs Test Scores
        iterations = np.arange(1, len(results['train_scores']) + 1)
        axes[0].errorbar(iterations, results['train_scores'], 
                         yerr=results['std_train_score'], 
                         label='Train', alpha=0.7, capsize=5)
        axes[0].errorbar(iterations, results['test_scores'], 
                         yerr=results['std_test_score'], 
                         label='Test', alpha=0.7, capsize=5)
        axes[0].set_xlabel('Iteration', fontsize=11)
        axes[0].set_ylabel('Accuracy', fontsize=11)
        axes[0].set_title('Train vs Test Accuracy', fontsize=12, fontweight='bold')
        axes[0].legend(fontsize=10)
        axes[0].grid(alpha=0.3)
        
        # Plot 2: Score Distribution
        axes[1].hist(results['train_scores'], bins=20, alpha=0.6, label='Train', edgecolor='black')
        axes[1].hist(results['test_scores'], bins=20, alpha=0.6, label='Test', edgecolor='black')
        axes[1].axvline(results['mean_train_score'], color='blue', linestyle='--', linewidth=2)
        axes[1].axvline(results['mean_test_score'], color='orange', linestyle='--', linewidth=2)
        axes[1].set_xlabel('Accuracy', fontsize=11)
        axes[1].set_ylabel('Frequency', fontsize=11)
        axes[1].set_title('Score Distribution', fontsize=12, fontweight='bold')
        axes[1].legend(fontsize=10)
        axes[1].grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        plt.show()


def _plot_regressor_results(results, cv_method):
    """
    Generate visualizations for regressor results
    """
    if cv_method in ['3way', 'kfold']:
        fig, ax = plt.subplots(figsize=(10, 6))
        folds = list(range(1, len(results['scores']) + 1))
        ax.bar(folds, results['scores'], alpha=0.7, color='green', edgecolor='black')
        ax.axhline(y=results['mean_score'], color='red', linestyle='--', linewidth=2, label=f"Mean R²: {results['mean_score']:.4f}")
        ax.fill_between(folds, 
                         results['mean_score'] - results['std_score'],
                         results['mean_score'] + results['std_score'],
                         alpha=0.2, color='red')
        ax.set_xlabel('Fold', fontsize=12)
        ax.set_ylabel('R² Score', fontsize=12)
        ax.set_title(f'Regression: {cv_method.upper()} Cross-Validation', fontsize=14, fontweight='bold')
        ax.set_xticks(folds)
        ax.legend(fontsize=10)
        ax.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()
    
    elif cv_method == 'montecarlo':
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        
        # Plot 1: Train vs Test R² Scores
        iterations = np.arange(1, len(results['train_scores']) + 1)
        axes[0].plot(iterations, results['train_scores'], 
                     marker='o', label='Train', alpha=0.7, markersize=4)
        axes[0].plot(iterations, results['test_scores'], 
                     marker='s', label='Test', alpha=0.7, markersize=4)
        axes[0].axhline(y=results['mean_train_score'], color='blue', linestyle='--', alpha=0.5)
        axes[0].axhline(y=results['mean_test_score'], color='orange', linestyle='--', alpha=0.5)
        axes[0].set_xlabel('Iteration', fontsize=11)
        axes[0].set_ylabel('R² Score', fontsize=11)
        axes[0].set_title('Train vs Test R² Score', fontsize=12, fontweight='bold')
        axes[0].legend(fontsize=10)
        axes[0].grid(alpha=0.3)
        
        # Plot 2: Score Distribution
        axes[1].hist(results['train_scores'], bins=20, alpha=0.6, label='Train', edgecolor='black')
        axes[1].hist(results['test_scores'], bins=20, alpha=0.6, label='Test', edgecolor='black')
        axes[1].axvline(results['mean_train_score'], color='blue', linestyle='--', linewidth=2)
        axes[1].axvline(results['mean_test_score'], color='orange', linestyle='--', linewidth=2)
        axes[1].set_xlabel('R² Score', fontsize=11)
        axes[1].set_ylabel('Frequency', fontsize=11)
        axes[1].set_title('R² Score Distribution', fontsize=12, fontweight='bold')
        axes[1].legend(fontsize=10)
        axes[1].grid(axis='y', alpha=0.3)
        
        plt.tight_layout()
        plt.show()
```

## Example Usage: Classification

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier

# Load iris dataset
iris = load_iris()
X_iris = iris.data
y_iris = iris.target

# Example 1: 3-Way Fold Classification
print("\n" + "="*60)
print("CLASSIFICATION EXAMPLE 1: 3-Way Fold")
print("="*60)
clf = KNeighborsClassifier(n_neighbors=5)
results_3way = automl_classifier(X_iris, y_iris, clf, cv_method='3way', plot=True)

# Example 2: K-Fold Classification
print("\n" + "="*60)
print("CLASSIFICATION EXAMPLE 2: 5-Fold Cross-Validation")
print("="*60)
clf = KNeighborsClassifier(n_neighbors=5)
results_kfold = automl_classifier(X_iris, y_iris, clf, cv_method='kfold', 
                                   cv_params={'n_splits': 5}, plot=True)

# Example 3: Monte Carlo Classification
print("\n" + "="*60)
print("CLASSIFICATION EXAMPLE 3: Monte Carlo Cross-Validation")
print("="*60)
clf = KNeighborsClassifier(n_neighbors=5)
results_mc = automl_classifier(X_iris, y_iris, clf, cv_method='montecarlo', 
                               cv_params={'n_iterations': 100}, plot=True)
```

## Example Usage: Regression

```python
from sklearn.datasets import make_regression
from sklearn.neighbors import KNeighborsRegressor

# Create synthetic regression dataset
X_reg, y_reg = make_regression(n_samples=200, n_features=10, noise=10, random_state=42)

# Example 1: 3-Way Fold Regression
print("\n" + "="*60)
print("REGRESSION EXAMPLE 1: 3-Way Fold")
print("="*60)
reg = KNeighborsRegressor(n_neighbors=5)
results_reg_3way = automl_regressor(X_reg, y_reg, reg, cv_method='3way', plot=True)

# Example 2: K-Fold Regression
print("\n" + "="*60)
print("REGRESSION EXAMPLE 2: 5-Fold Cross-Validation")
print("="*60)
reg = KNeighborsRegressor(n_neighbors=5)
results_reg_kfold = automl_regressor(X_reg, y_reg, reg, cv_method='kfold', 
                                      cv_params={'n_splits': 5}, plot=True)

# Example 3: Monte Carlo Regression
print("\n" + "="*60)
print("REGRESSION EXAMPLE 3: Monte Carlo Cross-Validation")
print("="*60)
reg = KNeighborsRegressor(n_neighbors=5)
results_reg_mc = automl_regressor(X_reg, y_reg, reg, cv_method='montecarlo', 
                                   cv_params={'n_iterations': 100}, plot=True)
```

## Quick Reference: Function Parameters

### automl_classifier() & automl_regressor()

**Required Parameters:**
- `X`: Feature matrix (array-like or DataFrame)
- `y`: Target variable (array-like or Series)
- `model`: Instantiated sklearn model object

**Optional Parameters:**
- `cv_method` (str): 'kfold', '3way', or 'montecarlo' (default: 'kfold')
- `cv_params` (dict): Cross-validation parameters
  - For 'kfold': `{'n_splits': 5}`
  - For 'montecarlo': `{'n_iterations': 100, 'test_size': 0.25}`
- `test_size` (float): Test split size for montecarlo (default: 0.25)
- `scale_features` (bool): Standardize features (default: True)
- `random_state` (int): Reproducibility seed (default: 42)
- `plot` (bool): Generate visualizations (default: False)

**Returns:**
- `results` (dict): Comprehensive results including scores, statistics, and trained model
